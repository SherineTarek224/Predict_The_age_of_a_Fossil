# **ENcoding**

```python
la=LabelEncoder()
df['geological_period']=la.fit_transform(df['geological_period'])
df['surrounding_rock_type']=la.fit_transform(df['surrounding_rock_type'])
df['paleomagnetic_data']=la.fit_transform(df['paleomagnetic_data'])
df['inclusion_of_other_fossils']=la.fit_transform(df['inclusion_of_other_fossils'])
df['stratigraphic_position']=la.fit_transform(df['stratigraphic_position'])
```

# **Training Model**

```python
x=df.drop(columns=['age'])
y=df['age']
```

```python
x.head()
```

```python
y.head()
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error

lr=LinearRegression()
lr.fit(x_train,y_train)
y_pred=lr.predict(x_test)
```

# **Evaluation**

```python
print("Linear regression cost function")
cost=mean_squared_error(y_test,y_pred)
print("cost",cost)
```
